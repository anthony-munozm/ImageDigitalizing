# Estimated Suppression for 1M USD LIBOR.csv

```python
name = "Estimated Suppression for 1M USD LIBOR.csv"
data = pd.read_csv(name,  header=None)
data = data[data[1]>0.05]
data = data[data[0]>1]
data = data[data[1]<2.45]
a = dt.datetime(2007,8,1,0,0,0)
b = dt.datetime(2010,5,1,0,0,0)


pd.set_option('display.float_format', '{:.2f}'.format)
data["date"] = pd.date_range(start='2007-8-1 00:00:00', end='2010-5-31 00:00:00', periods=len(data))
data["epoch"] = int(dt.datetime(2007,8,1,0,0,0).strftime('%s')) + data[0]*(b-a).total_seconds()/data[0].values[-1]
data['date_string'] = pd.to_datetime(data["epoch"], unit='s')

data["date_"] = pd.to_datetime(data['date_string']).dt.floor('d')
data = data.groupby(["date_"]).mean().reset_index()
data.plot.line(x="date_", y=1, figsize=(15,5))

data = data.set_index("date_")
data = data.reindex(pd.date_range(start='2007-8-1 00:00:00', end='2010-5-31 00:00:00', freq="D"), method="nearest")
data["identifier"] = name
data = data.reset_index()
data = data.rename(columns={1:"value", "index":"date"})
data= data[["value","identifier","date"]]
```

```python
General = data
```
